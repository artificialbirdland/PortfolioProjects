```python
# Import libraries

import pandas as pd
import numpy as np
from scipy.stats import zscore  
```

```python
# Import the dataset

df = pd.read_excel(r"C:\Users\Enes\Desktop\Project07\unirankings1.xlsx")
df
```

```python
# Dividing some columns to 10 because of an error occured while importing and exporting the data. (Not neccecary)

# Rank

df['Rank'] = df['Rank']/10
df['Rank'] = df['Rank'].astype(int)

# Students to Staff Ratio

df['Students to Staff Ratio'] = df['Students to Staff Ratio']/10

# Overall Score
df['Overall Score'] = df['Overall Score']/10
df['Overall Score'] = df['Overall Score'].apply(lambda x: '{:.1f}'.format(x))

# Teaching
df['Teaching'] = df['Teaching']/10
df['Teaching'] = df['Teaching'].apply(lambda x: '{:.1f}'.format(x))

# Research Environment
df['Research Environment'] = df['Research Environment']/10
df['Research Environment'] = df['Research Environment'].apply(lambda x: '{:.1f}'.format(x))

# Research Quality
df['Research Quality'] = df['Research Quality']/10
df['Research Quality'] = df['Research Quality'].apply(lambda x: '{:.1f}'.format(x))

# Industry Impact
df['Industry Impact'] = df['Industry Impact']/10
df['Industry Impact'] = df['Industry Impact'].apply(lambda x: '{:.1f}'.format(x))

# International Outlook
df['International Outlook'] = df['International Outlook']/10#df['International Outlook'] = df['International Outlook'].apply(lambda x: '{:.1f}'.format(x))

df
```

```python
# Shaping the Female to Male Ratio for further organizing data.

df['Female to Male Ratio'] = df['Female to Male Ratio'].str.replace(' ', '').str[:5]

df
```

```python
# Seperating and creating new columns for further analysis.

# Creating Local Student Population
df['Local Student Population'] = df['Student Population'] - (df['Student Population']*df['International Students'])

# Creating International Student Population
df['International Student Population'] = df['Student Population'] - df['Local Student Population']
df
```

```python
# Creating Local Female Student Population with help of using the first 2 digits of Female to Male Ratio
df['Local Female Student Population'] = df.apply(
    lambda row: row['Local Student Population'] * float(row['Female to Male Ratio'][:2]) / 100
    if pd.notna(row['Local Student Population']) and pd.notna(row['Female to Male Ratio']) else pd.NA, axis=1)

# We can create Local Male Student Population with substracting Local Female Student Pop. from Local Student Population 
df['Local Male Student Population'] = df['Local Student Population'] - df['Local Female Student Population']
df
```

```python
# Doing the same iteration here for International Students
df['International Female Student Population'] = df.apply(
    lambda row: row['International Student Population'] * float(row['Female to Male Ratio'][:2]) / 100
    if pd.notna(row['International Student Population']) and pd.notna(row['Female to Male Ratio']) else pd.NA, axis=1)

df['International Male Student Population'] = df['International Student Population'] - df['International Female Student Population']
df
```

```python
# Creating new column called Number of Staff for further analysis
df['Number of Staff'] = df['Student Population']/df['Students to Staff Ratio']
df
```

```python
# Convert string values in 'Teaching', 'Research Environment', 'Research Quality', 'Industry Impact', 'International Outlook' columns to numeric values
df[['Teaching', 'Research Environment', 'Research Quality', 'Industry Impact', 'International Outlook']] = df[['Teaching', 'Research Environment', 'Research Quality', 'Industry Impact', 'International Outlook']].apply(pd.to_numeric, errors='coerce')

# Calculating the 'Overall Score' for more accurate values 
df['Overall Score'] = (df['Teaching']+df['Research Environment']+df['Research Quality']+df['Industry Impact']+df['International Outlook'])/5
df
```

```python
# I looked at the dataset and most of the outliers are in 'Overall Score' column. So let's fix it!
# Specifying the column names from which you want to detect outliers
selected_columns = ['Overall Score']

# Detecting outliers using Z-Score
zscore_threshold = 3
outliers_zscore = df[selected_columns][(np.abs(zscore(df[selected_columns])) > zscore_threshold).any(axis=1)]

# Show the results of the method
print("Z-Score Outliers:")
print(outliers_zscore)
```

```python
# I did this proccess several times and '14'th index also had outliers but didn't show. So I added here too.
# With help of excel I added the outlier indexes here :D
outlier_scores = [14, 93, 116, 121, 165, 210, 212, 418, 422, 461, 475, 476, 494, 504, 530, 533, 546, 563, 602, 616, 652, 669, 678, 683, 699, 722, 735, 745, 749, 759, 764, 768, 769, 773, 779, 781, 785]

# Find rows with Outlier scores
result_df = df[df.index.isin(outlier_scores)]

result_df
```

```python
# I don't know why but in the middle of proccess the df wasn't showing me Industry Impact, so I wrote this code:
pd.set_option('display.max_columns', None)

# Filter rows with specific indexes
selected_rows = df[df.index.isin(outlier_scores)].copy()  # copy() ile yeni bir DataFrame oluştur

# Get average by selecting specific columns
average_value = selected_rows[['Teaching', 'Research Environment', 'Research Quality', 'International Outlook']].mean(axis=1)

# Update "Industry Impact" column with average
selected_rows.loc[:, 'Industry Impact'] = average_value

# Show synchronized DataFrame
selected_rows
```

```python
# Last proccess was about getting averages of 'Teaching', 'Research Environment', 'Research Quality', 'International Outlook' to Update 'Industry Impact'
# Now we have to get averages of all the values and update 'Overall Score'
# Get average by selecting specific columns
average_value = selected_rows[['Teaching', 'Research Environment', 'Research Quality', 'International Outlook','Industry Impact']].mean(axis=1)

# Update "Industry Impact" column with average
selected_rows.loc[:, 'Overall Score'] = average_value

# Show synced DataFrame
selected_rows
```
