```python
#Import the libraries
from bs4 import BeautifulSoup
import requests
```

```python
# This program extracts market data from the Wikipedia page on "Piyasa Değeri" (Market Capitalization).
# It uses the requests library to fetch the HTML content from the given URL and BeautifulSoup for parsing the HTML.
# The extracted data can be further processed and utilized as needed.
url = 'https://tr.wikipedia.org/wiki/Piyasa_de%C4%9Feri'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
# Print the BeautifulSoup object 'soup' to inspect the HTML structure obtained from the Wikipedia page.
print(soup)
```

```python
# Find the first 'table' element in the HTML content using BeautifulSoup.
table = soup.find('table')

# Print the 'table' element to inspect the structure of the HTML table on the Wikipedia page.
print(table)
```

```python
# Extract table headers (titles) from the HTML table on the Wikipedia page.
# Iterate over each row ('tr') in the table, find all header cells ('th'), and append the stripped text to the 'titles' list.
titles = []
for row in table.find_all('tr'):
    cells = row.find_all('th')
    for cell in cells:
        titles.append(cell.text.strip())

# Print the extracted titles for inspection.
print(titles)
```

```python
import pandas as pd
# Create an empty Pandas DataFrame with columns defined by the extracted titles from the HTML table.
df = pd.DataFrame(columns=titles)
# The DataFrame is initially empty and will be populated with data later in the code.
df
```

```python
# Extract all rows ('tr' elements) from the HTML table on the Wikipedia page.
column_data = table.find_all('tr')

# The 'column_data' variable now contains a list of all rows in the HTML table.
# Each row may contain data cells ('td') and/or header cells ('th').
column_data = table.find_all('tr')
print(column_data)
```

```python
# Iterate through each row (excluding the first row, which contains headers) in the HTML table.
# For each row, find all data cells ('td' elements) and extract the stripped text.
# Populate the Pandas DataFrame with the extracted row data.
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    # Get the length of the DataFrame for proper indexing and add the individual row data.
    length = len(df)
    df.loc[length] = individual_row_data

# Display the populated DataFrame.
df
```

```python
# Export the populated Pandas DataFrame to a CSV file.
# The 'piyasaddegeri.csv' file will be saved on the desktop in the 'deneme' folder.
# The file is encoded in UTF-8, and the index column is excluded from the CSV.
df.to_csv(r'C:\Users\Enes\Desktop\deneme\piyasaddegeri.csv', encoding = 'UTF8',index = False)
```
